```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

dataframe = pd.read_csv('Results.csv',header = 1 , low_memory = False)
print(dataframe)
```

Data Cleaning :

```python
dataframe.columns = ['summary_id', 'putaway_Date', 'count', 'item_Code', 'putAway_Qty', 
                     'ord_Qty', 'storageBin', 'storageSectionId', 'total_Time_in_Seconds', 
                     'perQty_Time', 'item_text']
dataframe['putaway_Date'] = pd.to_datetime(dataframe['putaway_Date'], format='%d-%m-%Y')
dataframe['total_Time_in_Seconds'] = pd.to_numeric(dataframe['total_Time_in_Seconds'], errors='coerce')
dataframe['perQty_Time'] = pd.to_numeric(dataframe['perQty_Time'], errors='coerce')
dataframe['putAway_Qty'] = pd.to_numeric(dataframe['putAway_Qty'], errors='coerce')
dataframe['ord_Qty'] = pd.to_numeric(dataframe['ord_Qty'], errors='coerce')
dataframe['count'] = pd.to_numeric(dataframe['count'], errors='coerce')
print(dataframe.head())
```

```python
print(dataframe.isnull().sum())
dataframe['item_text'] = dataframe['item_text'].fillna('Unknown_Item')
print('Changed Null Values to Unknown_Item')
dataframe['putaway_Date'] = pd.to_datetime(dataframe['putaway_Date'] , format = 'mixed' , dayfirst = True)
print("Dates Cleansed ")
```

Features Creation:

Feature 1  : PutAway ThroughPut

```python
''' PutAway Throughput is Volume successfully stored per unit of time 
Thus PutAway Throughput = putAway_Qty / total_Time_in_Seconds = 1 / perQty_Time'''

#threshold = float(input('Enter The Threshold Value : '))
threshold = 98.0
dataframe['putawayMonth'] = pd.to_datetime(dataframe['putaway_Date']).dt.to_period('M')

putaway_throughput_df = dataframe.groupby(['putawayMonth', 'storageSectionId']).agg(
    zone_units=('putAway_Qty', 'sum'),
    zone_time=('total_Time_in_Seconds', 'sum')
).reset_index()

putaway_throughput_df['putaway_throughput'] = (putaway_throughput_df['zone_units'] / (putaway_throughput_df['zone_time'] + 0.001))

best_zone_df = putaway_throughput_df.groupby('putawayMonth')['putaway_throughput'].max().reset_index()
best_zone_df.rename(columns={'putaway_throughput': 'best_putaway_throughput'}, inplace=True)

putaway_throughput_df = pd.merge(putaway_throughput_df, best_zone_df, on='putawayMonth', how='left')
putaway_throughput_df['Normalized_Score'] = (
    putaway_throughput_df['putaway_throughput'] / putaway_throughput_df['best_putaway_throughput']
) * 100

putaway_throughput_df['Zone_Score_Pass'] = putaway_throughput_df['Normalized_Score'] >= threshold
print(putaway_throughput_df)
'''
d = {}
for k in range(98, 89, -1):
    k = float(k)
    passes = (putaway_throughput_df['Normalized_Score'] >= k).sum()  # no need to modify df column
    d[f'{k}'] = passes

thresholds = list(d.keys())
values     = list(d.values())

plt.figure(figsize=(10, 6))

plt.plot(thresholds, values, marker='o', linewidth=2, markersize=6, color='steelblue')

if 90 <= threshold <= 98:
    thresh_key = f'{float(threshold)}'
    if thresh_key in d:
        plt.axvline(x=thresh_key, color='red', linestyle='--', linewidth=1.5, label=f'Your Threshold ({threshold})')
        plt.legend()

for x, y in zip(thresholds, values):
    plt.annotate(str(int(y)), (x, y), textcoords="offset points", xytext=(0, 8), ha='center', fontsize=9)

plt.xlabel('Thresholds')
plt.ylabel('Number Of True Cases')
plt.title('Threshold Estimator — Zones Passing at Each Normalized Score')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()'''
print()
```

Feature 3 : Average Drop Size 

```python
'''Median Drop Size is median(putAway_Qty) grouped by zone-month , separating Bulk Recieving Zones and Single Unit recieving ones '''

Use_Count = False

dataframe['storageSectionId'] = dataframe['storageSectionId'].str.upper().str.strip()

if Use_Count:
    dataframe['effective_drop_qty'] = dataframe['count'] * dataframe['putAway_Qty']
    qty_label = "count × putAway_Qty"
else:
    dataframe['effective_drop_qty'] = dataframe['putAway_Qty']
    qty_label = "putAway_Qty"

print(f"    Mode: {qty_label}")

zone_drop_df = dataframe.groupby(['putawayMonth', 'storageSectionId']).agg(
    median_drop_size=('effective_drop_qty', 'median'),
    mean_drop_size=('effective_drop_qty', 'mean'),
    total_transactions=('effective_drop_qty', 'count'),
    total_putaway_qty=('effective_drop_qty', 'sum')
).reset_index()

zone_drop_df['mean_median_ratio'] = (
    zone_drop_df['mean_drop_size'] / (zone_drop_df['median_drop_size'] + 0.001)
)

skew_flag_threshold = 5.0
zone_drop_df['Bulk_Spike_Flag'] = zone_drop_df['mean_median_ratio'] > skew_flag_threshold

best_drop_df = zone_drop_df.groupby('putawayMonth')['median_drop_size'].max().reset_index()
best_drop_df.rename(columns={'median_drop_size': 'best_median_drop'}, inplace=True)

zone_drop_df = pd.merge(zone_drop_df, best_drop_df, on='putawayMonth', how='left')

zone_drop_df['Avg_Drop_Size_Score'] = (
    zone_drop_df['median_drop_size'] / (zone_drop_df['best_median_drop'] + 0.001)
) * 100

q3_benchmark = zone_drop_df['Avg_Drop_Size_Score'].quantile(0.75)
q1_benchmark = zone_drop_df['Avg_Drop_Size_Score'].quantile(0.25)

if q1_benchmark == q3_benchmark:
    q1_benchmark = zone_drop_df['Avg_Drop_Size_Score'].median() * 0.5

def classify_drop_profile(score, q3_limit, q1_limit):
    if score >= q3_limit:
        return 'Bulk-Profile Zone'
    elif score >= q1_limit:
        return 'Standard-Profile Zone'
    else:
        return 'Micro-Drop Zone'

zone_drop_df['Drop_Profile'] = zone_drop_df['Avg_Drop_Size_Score'].apply(
    lambda x: classify_drop_profile(x, q3_benchmark, q1_benchmark)
)

print(f"\nDrop Size computed as: {qty_label}")
print(f"Skew Flag Threshold Value :(Suggested is 5)  {int(skew_flag_threshold)}")
print(f"\nWarehouse-Wide Median Drop Size Baseline: {zone_drop_df['median_drop_size'].median():.2f} units/transaction")
print(f"Calculated Q3 Bulk-Profile Benchmark: {q3_benchmark:.2f}")
print(f"Calculated Q1 Micro-Drop Benchmark:   {q1_benchmark:.2f}")

total_zone_months = len(zone_drop_df)
bulk_count = len(zone_drop_df[zone_drop_df['Drop_Profile'] == 'Bulk-Profile Zone'])
standard_count = len(zone_drop_df[zone_drop_df['Drop_Profile'] == 'Standard-Profile Zone'])
micro_count = len(zone_drop_df[zone_drop_df['Drop_Profile'] == 'Micro-Drop Zone'])
spike_count = zone_drop_df['Bulk_Spike_Flag'].sum()

print("\n" + "=" * 60)
print("PARAMETER AUDIT: AVERAGE DROP SIZE SUMMARY REPORT")
print("=" * 60)
print(f"Total Zone-Month combinations processed: {total_zone_months} rows")
print(f"Bulk-Profile Zones (>= Q3 {q3_benchmark:.2f}): {bulk_count} rows")
print(f"Standard-Profile Zones (Q1 to Q3): {standard_count} rows")
print(f"Micro-Drop Zones (< Q1 {q1_benchmark:.2f}): {micro_count} rows")
print(f"Bulk Spike Warnings (Mean/Median ratio > {skew_flag_threshold}): {spike_count} rows")
```

Feature 4 : Weighted Lead Time 

```python
dataframe['weighted_time_component'] = dataframe['total_Time_in_Seconds'] * dataframe['putAway_Qty']

weighted_lt_df = dataframe.groupby(['putawayMonth', 'storageSectionId']).agg(
    sum_weighted_time=('weighted_time_component', 'sum'),
    sum_putaway_qty=('putAway_Qty', 'sum'),
    total_transactions=('putAway_Qty', 'count')
).reset_index()

weighted_lt_df['Weighted_Lead_Time'] = (
    weighted_lt_df['sum_weighted_time'] / (weighted_lt_df['sum_putaway_qty'] + 0.001)
)

print("--- Putaway Parameter 4: Weighted Lead Time Score Engine ---")
print(f"Warehouse-Wide Median Weighted Lead Time: {weighted_lt_df['Weighted_Lead_Time'].median():.2f} seconds")

best_lt_df = weighted_lt_df.groupby('putawayMonth')['Weighted_Lead_Time'].min().reset_index()
best_lt_df.rename(columns={'Weighted_Lead_Time': 'best_lead_time'}, inplace=True)

weighted_lt_df = pd.merge(weighted_lt_df, best_lt_df, on='putawayMonth', how='left')

weighted_lt_df['Weighted_Lead_Time_Score'] = (
    weighted_lt_df['best_lead_time'] / (weighted_lt_df['Weighted_Lead_Time'] + 0.001)
) * 100

q3_benchmark = weighted_lt_df['Weighted_Lead_Time_Score'].quantile(0.75)
q1_benchmark = weighted_lt_df['Weighted_Lead_Time_Score'].quantile(0.25)

if q1_benchmark == q3_benchmark:
    q1_benchmark = weighted_lt_df['Weighted_Lead_Time_Score'].median() * 0.5

def classify_lead_time_profile(score, q3_limit, q1_limit):
    if score >= q3_limit:
        return 'Fast Zone'
    elif score >= q1_limit:
        return 'Standard Zone'
    else:
        return 'Slow Zone'

weighted_lt_df['Lead_Time_Profile'] = weighted_lt_df['Weighted_Lead_Time_Score'].apply(
    lambda x: classify_lead_time_profile(x, q3_benchmark, q1_benchmark)
)

print(f"\nCalculated Q3 Fast-Zone Benchmark:     {q3_benchmark:.2f}")
print(f"Calculated Q1 Slow-Zone Benchmark:     {q1_benchmark:.2f}")

total_zone_months = len(weighted_lt_df)
fast_count = len(weighted_lt_df[weighted_lt_df['Lead_Time_Profile'] == 'Fast Zone'])
standard_count = len(weighted_lt_df[weighted_lt_df['Lead_Time_Profile'] == 'Standard Zone'])
slow_count = len(weighted_lt_df[weighted_lt_df['Lead_Time_Profile'] == 'Slow Zone'])

print("\n" + "=" * 60)
print("PARAMETER AUDIT: WEIGHTED LEAD TIME SUMMARY REPORT")
print("=" * 60)
print(f"Total Zone-Month combinations processed: {total_zone_months} rows")
print(f"Fast Zones (>= Q3 {q3_benchmark:.2f}): {fast_count} rows")
print(f"Standard Zones (Q1 to Q3): {standard_count} rows")
print(f"Slow Zones (< Q1 {q1_benchmark:.2f}): {slow_count} rows")

print("\n--- Weighted Lead Time Matrix (First 20 Rows) ---")
columns_to_show = [
    'putawayMonth', 'storageSectionId', 'Weighted_Lead_Time',
    'Weighted_Lead_Time_Score', 'Lead_Time_Profile'
]
print(weighted_lt_df[columns_to_show].head(20).to_string(index=False))
```

Merging the Columns :

```python
pa1 = putaway_throughput_df[['putawayMonth', 'storageSectionId', 'Normalized_Score']].copy()
pa1.rename(columns={'Normalized_Score': 'Putaway_Throughput_Score'}, inplace=True)

pa3 = zone_drop_df[['putawayMonth', 'storageSectionId', 'Avg_Drop_Size_Score']].copy()

pa4 = weighted_lt_df[['putawayMonth', 'storageSectionId', 'Weighted_Lead_Time_Score']].copy()

for df in [pa1,pa3 , pa4]:
    df['putawayMonth'] = df['putawayMonth'].astype(str)
    df['storageSectionId'] = df['storageSectionId'].str.upper().str.strip()

putaway_master_df = pa1.merge(pa3, on=['putawayMonth', 'storageSectionId'], how='left')
putaway_master_df = putaway_master_df.merge(pa4, on=['putawayMonth', 'storageSectionId'], how='left')

putaway_kmeans_features = [
    'Putaway_Throughput_Score',
    'Completion_Rate',
    'Avg_Drop_Size_Score'
]

print(f"\nFeatures going into K-Means ({len(putaway_kmeans_features)}):")
for f in putaway_kmeans_features:
    print(f)
```

Scaling : 

```python
X = putaway_master_df[putaway_kmeans_features].fillna(0)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

K value Estimator : 

```python
from sklearn.metrics import silhouette_score
inertia = []
sil_scores = []
K_range = range(2, 7)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=67, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertia.append(km.inertia_)
    sil_scores.append(silhouette_score(X_scaled, labels))

fig, ax1 = plt.subplots(figsize=(10, 10))
ax1.plot(K_range, inertia, marker='o', color='skyblue', label='Inertia (Elbow)')
ax1.set_xlabel('Number of Clusters (k)', fontweight='bold')
ax1.set_ylabel('Inertia (Sum of Squared Distances)', color='blue', fontweight='bold')

ax2 = ax1.twinx()
ax2.plot(K_range, sil_scores, marker='s', color='green', label='Silhouette Score')
ax2.set_ylabel('Silhouette Score', color='green', fontweight='bold')

plt.title("Diagnostic: Finding the Optimal Number of Operational Tiers", fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
